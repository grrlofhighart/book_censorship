## Book Analysis

```python
import numpy as np
import pandas as pd

# sqlite3 for storage of the dataset
import sqlite3
from contextlib import closing
```

```python
# Setup SQLite Database to store data
database = 'Book_DB.db'
conn = sqlite3.connect(database)
c = conn.cursor()
```

```python
# Find books on el_list that are also on ALA list
query1 = '''
SELECT * FROM el_list
WHERE el_list.title IN
(
SELECT DISTINCT ala_top_10_archive_clean.title
FROM ala_top_10_archive_clean
JOIN el_list
WHERE ala_top_10_archive_clean.title=el_list.title COLLATE NOCASE
)
ORDER BY el_list.title;
'''
#seting the dataframe width to max
pd.set_option('display.max_colwidth', None)
#running our query
pd.read_sql_query(query1, conn)
```

```python
# Find books on el_list that are on ALA list where years match
query2 = '''
SELECT * FROM el_list
JOIN ala_top_10_archive_clean
WHERE el_list.title=ala_top_10_archive_clean.title
AND el_list.year=ala_top_10_archive_clean.year
'''
pd.read_sql_query(query2, conn)
```

```python
query3 = '''
SELECT * FROM el_list
JOIN ala_top_10_archive_clean
JOIN Public_State_Level_Data
WHERE el_list.state=Public_State_Level_Data.abbr
AND el_list.title=ala_top_10_archive_clean.title
AND el_list.year=ala_top_10_archive_clean.year
GROUP BY el_list.ID
'''
pd.read_sql_query(query3, conn)
```

```python
# Find books on el_list that are on ALA list with matching years
# Include geo and statistical data
query4 = '''
SELECT * FROM el_list
JOIN ala_top_10_archive_clean
JOIN Public_State_Level_Data
WHERE el_list.title=ala_top_10_archive_clean.title
AND el_list.year=ala_top_10_archive_clean.year
AND el_list.state IN
(
SELECT Public_State_Level_Data.abbr 
FROM Public_State_Level_Data
JOIN el_list
WHERE el_list.state=Public_State_Level_Data.abbr
)
GROUP BY el_list.ID
'''
pd.read_sql_query(query4, conn)
```

```python
# Create table of all books, schools from ala, el_list, public state level data
query5 = '''
CREATE TABLE mapped_bans
AS
SELECT * FROM el_list
JOIN ala_top_10_archive_clean
JOIN Public_State_Level_Data
WHERE el_list.state=Public_State_Level_Data.abbr
AND el_list.title=ala_top_10_archive_clean.title
AND el_list.year=ala_top_10_archive_clean.year
GROUP BY el_list.ID
'''
```

```python
# Table of ban data by race
query6 = '''
CREATE TABLE bansbyrace
AS
SELECT *
FROM Enrollments_by_Race_Ethnicity
JOIN mapped_bans_final
WHERE Enrollments_by_Race_Ethnicity.school LIKE mapped_bans_final.NAME COLLATE NOCASE
GROUP BY mapped_bans_final.ID;
'''
```

```python
# Table of ban data by grade
query7 = '''
CREATE TABLE bansbygrade
AS
SELECT *
FROM Enrollments_by_Grade
JOIN mapped_bans_final
WHERE Enrollments_by_Grade.school LIKE mapped_bans_final.NAME COLLATE NOCASE
GROUP BY mapped_bans_final.ID;
'''
```
